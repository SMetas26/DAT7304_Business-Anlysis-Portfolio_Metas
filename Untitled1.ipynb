```python
#Importing libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
mean_absolute_error,
mean_squared_error,
mean_absolute_percentage_error
)
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("hotel_bookings.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

1.Removing Duplicate Values

```python
df = df.drop_duplicates()
```

```python
print(df.shape)
```

2.Handling Missing values 

```python
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
```

```python
df.isnull().sum()
```

```python
#creating total guests variables

df['total_guests'] = (
    df['adults']
    + df['children']
    + df['babies']
)

df[['adults','children','babies','total_guests']].head()
```

```python
#Total stay duration 

df['total_nights'] = (
    df['stays_in_week_nights']
    + df['stays_in_weekend_nights']
)

df[['stays_in_week_nights',
    'stays_in_weekend_nights',
    'total_nights']].head()
```

```python
#Estimated Revenue
df['revenue'] = (
    df['adr']
    *df['total_nights']
)

df[['adr','total_nights','revenue']].head()
```

```python
#Create arrival date

df['arrival_date'] = pd.to_datetime(
    df['arrival_date_day_of_month'].astype(str)
    + '-'
    + df['arrival_date_month']
    + '-'
    + df['arrival_date_year'].astype(str),
    format='%d-%B-%Y'
)

df[['arrival_date']].head()
```

```python
#Time series preparation

monthly_bookings = df.resample(
    'M',
    on='arrival_date'
).size()

monthly_bookings.head()
```

```python
#Train/Test Split
train = monthly_bookings.iloc[:-6]
test = monthly_bookings.iloc[-6:]
print("Train:", train.index.min(), "to", train.index.max())
print("Test:", test.index.min(), "to", test.index.max())
```

```python
def evaluate_forecast(actual, forecast, model_name):

    mae = mean_absolute_error(actual, forecast)

    rmse = np.sqrt(
    mean_squared_error(actual, forecast)
    )

    mape = mean_absolute_percentage_error(
    actual,
    forecast
    ) * 100

    results = pd.DataFrame({
    'Model':[model_name],
    'MAE':[mae],
    'RMSE':[rmse],
    'MAPE (%)':[mape]
    })
    
    return results 
```

```python
monthly_bookings.plot(figsize=(12,6))

plt.title('Monthly Hotel Bookings')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')

plt.show()
```

```python
#Naive Forecast
naive_forecast = np.repeat(
    train.iloc[-1],
    len(test)
)

plt.figure(figsize=(12,6))

plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(
    test.index,
    naive_forecast,
    label='Naive Forecast'
)

plt.legend()
plt.title('Naive Forecast')
plt.show()

evaluate_forecast(
    test,
    naive_forecast,
    'Naive Forecast'
)
```

```python
#Seasonal Naive Forecast

seasonal_naive = train.iloc[-6:].values

plt.figure(figsize=(12,6))

plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(
    test.index,
    seasonal_naive,
    label='Seasonal Naive'
)

plt.legend()
plt.title('Seasonal Naive Forecast')
plt.show()

evaluate_forecast(
    test,
    seasonal_naive,
    'Seasonal Naive'
)
```

```python
#Stationery Test/ADF
result = adfuller(monthly_bookings)

print('ADF Statistics:', result[0])
print('p-value:', result[1])
```

```python
monthly_diff = monthly_bookings.diff().dropna()
```

```python
result = adfuller(monthly_diff)

print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
#ACF and PACF
plot_acf(
    monthly_diff,
    lags=20
)

plt.show()
```

```python
plot_pacf(
    monthly_diff,
    lags=10
)

plt.show()
```

```python
#ARIMA Modelling
arima_model = ARIMA(
    train,
    order=(1,1,1)
).fit()
```

```python
print(arima_model.summary())
```

```python
#Forecast
arima_forecast = arima_model.forecast(
    steps=len(test)
)
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    train,
    label='Train'
)

plt.plot(
    test,
    label='Actual'
)

plt.plot(
    test.index,
    arima_forecast,
    label='Forecast'
)
plt.legend()

plt.show()
```

```python
mae = mean_absolute_error(
    test,
    arima_forecast
)
rmse = np.sqrt(
    mean_squared_error(
        test,
        arima_forecast
    )
)

print ("MAE:", mae)
print ("RMSE:", rmse)
```

```python
mape = mean_absolute_percentage_error(
    test,
    arima_forecast
) * 100
```

```python
print ("MAE:", mae)
print("RMSE:", rmse)
print("MAPE:", mape)
```

```python
#Residual Diagnostics
residuals = arima_model.resid

plt.figure(figsize=(12,6))

plt.plot(residuals)
plt.title('ARIA Residuals')
plt.show()
```

```python
plot_acf(
    residuals.dropna(),
    lags=12
)

plt.title('ACF of ARIMA Residuals')

plt.show()
```

```python
#Comparing Forecast
results = pd.concat([
    evaluate_forecast(
        test,
        naive_forecast,
        'Naive'
    ),

    evaluate_forecast(
        test,
        seasonal_naive,
        'Seasonal Naive'
    ),

    evaluate_forecast(
        test,
        arima_forecast,
        'ARIMA(1,1,1)'
    )
], ignore_index=True)
results.sort_values(
    by='RMSE'
)
```
